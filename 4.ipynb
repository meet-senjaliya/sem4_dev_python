```python
import plotly.express as px
import pandas as pd
data = {
    'Department':['IT','IT','HR','HR','Fin','Fin'],
    'Gender':['Male','Female','Male','Female','Male','Female'],
    'salary':[50000,55000,45000,48000,70000,72000]
}
df = pd.DataFrame(data)
print(df)
```

```python
# Create a bar chart
fig = px.bar(df, x="Department", y="salary", color="Gender", barmode="group")

# Show the chart
fig.show()
```

```python
import pandas as pd
df = pd.read_csv("social_media_engagement.csv")
df.head()
```

```python
fig = px.bar(df,
            x='platform',
            y='likes',
            color='post_type',
            pattern_shape='post_type',
            hover_data=['shares','comments'],
            barmode='relative')

fig.show()
```

```python
fig = px.scatter_matrix(df,
                       dimensions=['likes','comments','shares'],
                       color='likes',
                       title='Scatter Matrix')
fig.show()
```

```python
fig = px.scatter(df,
                x='likes',y='comments',color='platform',width=900,height=500,title='social media analyst',color_discrete_sequence=['blue','red','cyan'],
                facet_col='platform',
                facet_row='sentiment_score')
fig.show()
```

```python
fig=px.pie(df,
          values='sales',
          name='month',
          colour_discrete_sequence=['red'])
```
